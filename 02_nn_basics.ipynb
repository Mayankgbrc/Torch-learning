```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x = torch.tensor([[1.0, 2.0], [2.0, 3.0], [3.0, 4.0], [4.0, 5.0]])
y = torch.tensor([[8.0], [13.0], [18.0], [23.0]]) 
```

```python
class NNtest(nn.Module):
    def __init__(self):
        super(NNtest, self).__init__()
        # self.layer1 = nn.Linear(2,3)
        # self.layer2 = nn.Linear(3,2)
        self.layer3 = nn.Linear(2,1)
    def forward(self, x):
        # x = self.layer1(x)
        # x = torch.relu(x)
        # x = self.layer2(x)
        # x = torch.relu(x)
        x = self.layer3(x)
        print(x)
        return x
```

```python
model = NNtest()

loss_fn = nn.MSELoss()

optimizer = optim.SGD(model.parameters(), lr = 0.001)

print(model.parameters())

epochs = 100

for i in range(1, epochs + 1):
    predictions = model(x)

    loss = loss_fn(predictions, y)

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()

    if i % 20 == 0:
        print(f"epoch: {i}, loss: {loss}")
```
